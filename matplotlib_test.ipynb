```python
xdata = [1,2,3,4,5]
ydata = [1,3,5,7,9]

# plt.plot(xdata, ydata, color='red', linestyle='--', marker='o', markersize=10)
# plt.plot(xdata, ydata, color='blue', linestyle=':', marker='s', markersize=10)
plt.plot(xdata, ydata, color='c', linestyle='-.', marker='v', markersize=5)

plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(0, 6)  # x축의 범위를 0부터 6까지 설정
plt.ylim(0, 10)  # y축의 범위를 0부터 10까지 설정
plt.xticks(np.arange(0, 7, 1))  # x축의 눈금을 0부터 6까지 1씩 증가하도록 설정
plt.yticks(np.arange(0, 11, 1))  # y축의 눈금을 0부터 10까지 1씩 증가하도록 설정
plt.axhline(y=5, color='r', linestyle='--')  # y=5에서 수평선 추가
plt.axvline(x=3, color='g', linestyle='--')  # x=3에서 수직선 추가
plt.grid() # 격자 추가
plt.show()
```

```python
data = df 
data4 = data.loc[['부산', '대구', '인천', '경기'], 2020:2024]
data4
```

```python
plt.figure(figsize=(15, 4))  # 그래프 크기 설정
index = np.arange(4)
for year in range(2020, 2024):
    chartdata = data4[year]
    plt.bar(index, chartdata, label=f'{year}년', width=0.2)
    index = index + 0.2
plt.title('2020~2024년 미세먼지 농도')
plt.xticks(index-0.5 , ['부산', '대구', '인천', '경기'])
plt.show()
```

```python
# 누적막대로 그려보기
plt.figure(figsize=(15, 4))  # 그래프 크기 설정
index = np.arange(4)
left = np.zeros(4)
for year in range(2020, 2024):
    chartdata = data4[year]
    plt.barh(index, chartdata, left = left, label=f'{year}년')
    left = left + chartdata

plt.xlim(25, 180)
plt.ylabel('area')
plt.xlabel('micrometer')
plt.yticks(index , ['부산', '대구', '인천', '경기'])
plt.title('2020~2024년 미세먼지 농도')
plt.legend()
plt.show()
```

```python
data4.plot(kind='bar', figsize=(10,4), stacked=True)
data4.plot(kind='barh', figsize=(10,4), stacked=True)
```

```python
# 누적막대로 그려보기
plt.figure(figsize=(10, 4))  # 그래프 크기 설정
index = np.arange(4)
bottom = np.zeros(4)
for year in range(2020, 2024):
    chartdata = data4[year]
    plt.bar(index, chartdata, bottom = bottom, label=f'{year}년')
    bottom = bottom + chartdata

plt.ylim(25, 180)
plt.xlabel('area')
plt.ylabel('micrometer')
plt.xticks(index , ['부산', '대구', '인천', '경기'])
plt.title('2020~2024년 미세먼지 농도')
plt.legend()
plt.show()
```

```python
# 가로막대로 그려보기
plt.figure(figsize=(15, 4))  # 그래프 크기 설정
index = np.arange(4)
for year in range(2020, 2024):
    chartdata = data4[year]
    plt.barh(index, chartdata, height=0.2, label=f'{year}년')
    index = index + 0.2

plt.xlim(25, 50)
plt.ylabel('area')
plt.xlabel('micrometer')
plt.yticks(index-0.5 , ['부산', '대구', '인천', '경기'])
plt.title('2020~2024년 미세먼지 농도')
plt.legend()
plt.show()
```

```python
# 산점도 그려보기
health = pd.read_excel('data/health_screenings_2020_1000ea.xlsx', index_col=0)
health.head()
```

```python
health.info()
```

```python
heightdata = health['height']
weightdata = health['weight']
```

```python
plt.figure(figsize=(10, 4))  # 그래프 크기 설정
plt.scatter(heightdata, weightdata, alpha=0.5)  # alpha: 투명도
plt.title('키와 몸무게의 관계')
plt.xlabel('키')
plt.ylabel('몸무게')    
plt.grid()
plt.show()
```

```python
hw_df = pd.DataFrame({'height': heightdata, 'weight': weightdata})
hw_df.corr(method='pearson')  # 상관계수 계산
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
plt.figure(figsize=(15, 5))  # 그래프 크기 설정
plt.scatter(x, y)
plt.title('산점도')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.grid()
plt.show()
```

```python
systolic_data = health['systolic']
diastolic_data = health['diastolic']
blood_sugar_data = health['blood_sugar']

plt.figure(figsize=(10, 6))  # 그래프 크기 설정

plt.scatter(systolic_data, diastolic_data, color='r', edgecolors='w', label='systolic*diastolic')
plt.scatter(systolic_data, blood_sugar_data, color='g', edgecolors='w', label='systolic*blood_sugar')
plt.scatter(diastolic_data, blood_sugar_data, color='b', edgecolors='w', label='diastolic*blood_sugar')

plt.xlim(40,180)
plt.ylim(30,300)
plt.title('2020 Health Screening Group Scatter Graph')
plt.legend()
plt.grid()
plt.show()
```

```python
print(health['systolic'].corr(health['diastolic']))  # 상관계수 계산
print(health['systolic'].corr(health['blood_sugar']))  # 상관계수 계산
print(health['diastolic'].corr(health['blood_sugar']))  # 상관계수 계산
```

```python
# 히스토그램 그려보기

male_data = health.loc[health.gender == 1, ['gender', 'height']]
female_data = health.loc[health.gender == 2, ['gender', 'height']]

plt.figure(figsize=(10, 6))  # 그래프 크기 설정
plt.hist(female_data['height'], bins=9, alpha=0.5, color='r', label='Female')
plt.hist(male_data['height'], bins=9, alpha=0.5, color='b', label='Male')
plt.title('키 분포')
plt.xlabel('키')
plt.ylabel('인원수')
plt.legend()
plt.show()
```

```python
# 히스토그램 그려보기

male_data = health.loc[health.gender == 1, ['gender', 'weight']]
female_data = health.loc[health.gender == 2, ['gender', 'weight']]

plt.figure(figsize=(10, 6))  # 그래프 크기 설정
plt.hist(female_data['weight'], bins=10, alpha=0.5, color='r', label='Female')
plt.hist(male_data['weight'], bins=10, alpha=0.5, color='b', label='Male')
plt.title('몸무게 분포')
plt.xlabel('몸무게')
plt.ylabel('인원수')
plt.legend()
plt.show()
```

```python
# Box Plot(상자수염) 그려보기

weight = np.array([male_data['weight'], female_data['weight']], dtype=object)

plt.figure(figsize=(8, 5))  # 그래프 크기 설정
plt.boxplot(weight, label=['Male', 'Female'])
plt.ylim(20, 130)
plt.xlabel('gender')
plt.ylabel('weight')
plt.title('몸무게 분포')
plt.grid()
plt.show()
```

```python
# 가로로 Box Plot(상자수염) 그려보기

weight = np.array([male_data['weight'], female_data['weight']], dtype=object)

plt.figure(figsize=(8, 5))  # 그래프 크기 설정
plt.boxplot(weight, label=['Male', 'Female'], vert=False)
plt.xlim(20, 130)
plt.ylabel('gender')
plt.xlabel('weight')
plt.title('몸무게 분포')
plt.grid()
plt.show()
```
