```python
df=pd.DataFrame({'date':['2019-01-06', '2021-01-13', '2021-01-20', '2021-01-27', '2021-02-03'],
                  'x1':[5,4,3,2,1]})
df['date']=pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
df
```

```python
# rolling : 이동평균 구함(자기보다 날짜가 앞인 데이터를 가지고 계산함)
print('원본')
print('=='*20)
print(df)
temp=df.rolling(1).mean()
print('=='*20)
print(temp)
temp2=df.rolling(2).mean()
print('=='*20)
print(temp2)
temp3=df.rolling(3).mean()
print('=='*20)
print(temp3)
```

```python
# diff(앞데이터와의 차이값을 구해줌 )
print(df)
print('=='*20)
y_dff=df.diff()
print(y_dff)
```

```python
# 두 데이터를 다른 행으로 합침
temp = pd.concat([df, y_dff], axis=1)
temp
```

```python
temp2 = pd.concat([df, y_dff])
# temp2 = pd.concat([df, y_dff], axis=0)
# 위 두 코드는 같은 코드임. 두 값을 같은 행으로 합침(필드가 같을때만 가능함)
temp2
```

```python
print(df)
df['shift'] = df['x1'].shift(1)
print(df)
df['shift2'] = df['x1'].shift(2)
print(df)
```

```python
# bfill : 뒤쪽값으로 결측치를 채움
# ffill : 앞쪽값으로 결측치를 채움
df= df.fillna(method='bfill')
df
```

```python
df = fdr.DataReader('TSLA', start='2024-01-01')
df
```

```python
plt.Figure(figsize=(10,6))
sns.lineplot(x=df.index, y=df['Close'])
plt.title('TSLA Stock Price int 2024~')
plt.ylabel('Price')
plt.show()
```

```python
df.info()
```

```python
tsla_month = df.resample('BME').mean()
tsla_month 
```

```python
tsla_month['rtn']=tsla_month['Close'].pct_change()
tsla_month
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=tsla_month.index, y=tsla_month.rtn)
```

```python
tsla_month['MAE'] = tsla_month['Close'].rolling(2).mean()
tsla_month
```

```python
tsla_month.iloc[:,[3,7]].plot(figsize=(15,6))
```

```python
tsla_month['MAE2'] = tsla_month['Close'].rolling(3).mean()
tsla_month.iloc[:,[3,7,8]].plot(figsize=(15,6))
```

```python
last_close=tsla_month['MAE'].iloc[-2]
print(last_close)
price = tsla_month['Close'].iloc[-1]
print(price)
if price > last_close:
    print('상승장')
elif price<last_close:
    print('하락장')
else:
    print('변화없음')
```

ARIMA

```python
#ARIMA모델 만들기 
from statsmodels.tsa.arima.model import ARIMA
import statsmodels.api as sm
```

```python
model=ARIMA(tsla_month['Close'].values, order=(0,1,2))
model_fit =model.fit()
model_fit.summary()
```

```python
predict = model_fit.predict()
predict
```

```python
plt.figure(figsize=(10,6))
index=range(17)
plt.plot(index, tsla_month['Close'], label='Actual')
plt.plot(index, predict, label='Redict', linestyle='--', color='r')
```

```python
forecats_data = model_fit.forecast(steps=5)
print(forecats_data)
```
