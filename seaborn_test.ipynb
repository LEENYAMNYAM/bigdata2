```python
data6 = health.loc[:,['gender', 'height', 'weight', 'waist', 'drinking', 'smoking']]
data6.head()
```

```python
data6.loc[data6['gender'] == 1, ['gender']] = 'Male' # 성별 1을 남자로 변경
data6.loc[data6['gender'] == 2, ['gender']] = 'Female' # 성별 2를 여자로 변경
data6.loc[data6['drinking'] == 0, ['drinking']] = 'Non-drinking' # 음주 0을 음주로 변경
data6.loc[data6['drinking'] == 1, ['drinking']] = 'Drinking' # 음주 1를 비음주로 변경
data6.loc[data6['smoking'] == 1, ['smoking']] = 'Non-smoking' # 흡연 0를 비흡연으로 변경
data6.loc[data6['smoking'] == 2, ['smoking']] = 'Non-smoking' # 흡연 1를 비흡연으로 변경
data6.loc[data6['smoking'] == 3, ['smoking']] = 'Smoking' # 흡연 2을 흡연으로 변경
data6.head() # 변경된 데이터셋 출력
```

```python
drinking = data6.groupby(['gender', 'drinking'])['drinking'].count()
smoking = data6.groupby(['gender', 'smoking'])['smoking'].count()
drinking = drinking.to_frame(name='count')
smoking = smoking.to_frame(name='count')
drinking.reset_index(inplace=True)
smoking.reset_index(inplace=True)
print(drinking)
print(smoking)
```

```python
fig = plt.figure(figsize=(17,6))
area1=fig.add_subplot(1,2,1)   #subplot 1행 2열의 첫번째, 영역설정 
area2=fig.add_subplot(1,2,2)   
ax1 = sns.barplot(data=drinking, x='gender', y='count', hue='drinking', ax=area1)
ax2 = sns.barplot(data=smoking, x='gender', y='count', hue='smoking', ax=area2)
fig.suptitle('성별 음주 및 흡연 현황', fontsize=20)
area1.set_title('음주 현황', fontsize=15)
area2.set_title('흡연 현황', fontsize=15)
plt.show
```

```python
male_data = data6.loc[data6.gender =='Male', ['gender', 'weight', 'waist', 'drinking', 'smoking']]
female_data = data6.loc[data6.gender =='Female', ['gender', 'weight', 'waist', 'drinking', 'smoking']]
plt.figure(figsize=(10,5))
plt.title('Seaborn Strip Plot Graph')
sns.stripplot(data=male_data, x='waist', y='weight', color='blue')
sns.stripplot(data=female_data, x='waist', y='weight', color='orange')
plt.xticks(np.arange(0, 127, 63), labels=[53,90.5,128])
plt.show
# stripplot을 두번사용할 때 중복값이 생기면 지금처럼 에러가 발생할 수 있음.
```

```python
combined_data = pd.concat([male_data, female_data])
plt.figure(figsize=(10,5))
plt.title('Seaborn Strip Plot Graph')
sns.stripplot(data=combined_data, x='waist', y='weight', hue='gender', palette='Set2')
plt.xticks(np.arange(0, 127, 63), labels=[53,90.5,128])
plt.show()
```

```python
male_data_100 = male_data.head(100)
female_data_100 = female_data.head(100)
combined_data_100 = pd.concat([male_data_100, female_data_100])

plt.figure(figsize=(10,5))
plt.title('Seaborn Strip Plot Graph')
sns.swarmplot(data=combined_data_100, x='waist', y='weight', hue='gender', palette='dark', size=4)
plt.xticks(np.arange(0,46,22.5), labels=[53,75.5,98])
plt.show
```

```python
plt.Figure(figsize=(10,6))
sns.histplot(male_data['weight'], bins=7, label='Male', color='blue', alpha=0.5, kde=True)
sns.histplot(female_data['weight'], bins=7, label='Female', color='red', alpha=0.5, kde=True)
plt.legend()
plt.title('성별 체중 분포')
plt.xlabel('체중')
plt.ylabel('빈도수')
plt.grid()
plt.savefig('data/weight.png', dpi=300)
plt.show()
```
