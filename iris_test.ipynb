```python
import pandas as pd
titanic = pd.read_csv('data/train.csv')
titanic.head()
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
titanic.sort_values('Fare', ascending=False)
```

```python
pd.DataFrame(titanic['Survived'].value_counts())
```

```python
titanic.isnull().sum()
```

```python
titanic.drop(['Cabin'], axis=1, inplace=True)  # 열을 삭제 (행은 axis=0) inplace=True = 원본에 바로 적용
titanic.columns
```

```python
titanic['Embarked'].value_counts() # 승객들이 어디서 탔는지 종류별로 몇 명인지 출력
```

```python
titanic['Embarked'] = titanic['Embarked'].fillna('S') # 결측값을 'S'로 대체
titanic.isnull().sum() # 결측값이 없어진 것을 확인
```

```python
avg = titanic['Age'].mean() # 평균 나이
titanic['Age'] = titanic['Age'].fillna(avg) # 결측값을 평균 나이로 대체
```

```python
titanic['Pclass'].groupby(titanic["Embarked"]).mean() # Pclass의 평균을 Embarked별로 구하기
```

```python
titanic.groupby(["Pclass", "Sex"])[['Age','Survived']].mean()
```

```python
import seaborn
sns.countplot(x='Pclass', hue='Survived', data=titanic) # 승객 등급(Pclass)별로 몇 명이 탔고, 그중에 몇 명이 생존했는지 막대 그래프로 표현
```

```python
data=titanic.corr(numeric_only=True) # 문자나 문자열 있는 열은 무시하고 숫자열만 골라서
sns.heatmap(data=data, annot=True,cmap='Blues')
```
