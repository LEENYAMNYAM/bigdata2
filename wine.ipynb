```python
from scipy import stats                         # t-검정을 위한 라이브러리
from statsmodels.formula.api import ols, glm    # 회귀분석 라이브러리

red_wine_quality = wine.loc[wine['type'] =='red', 'quality']
red_wine_quality
```

```python
white_wine_quality = wine.loc[wine['type'] == 'white', 'quality']
white_wine_quality
```

```python
stats.ttest_ind(red_wine_quality, white_wine_quality, equal_var=False)
```

```python
Rformula='quality~fixed_acidity+volatile_acidity+citric_acid+residual_sugar+chlorides+free_sulfur_dioxide+total_sulfur_dioxide+density+pH+sulphates+alcohol'
# Rformula
regression_result=ols(Rformula, data=wine).fit()
```

```python
regression_result.summary()
```

```python
sample1 = wine[wine.columns.difference(['quality', 'type'])]   # wine에서 quality와 type 칼럼을 제외한 나머지 칼럼들을 sample1에 저장
sample2 = sample1[0:5][:]
sample2
```

```python
predict1 = regression_result.predict(sample2)
predict1
```

```python
wine[0:5]['quality']
```

```python
sample3 = sample1.tail()
predict2 = regression_result.predict(sample3)
predict2
```
