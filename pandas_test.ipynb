```python
df = pd.read_csv('data/weather.csv', encoding='ansi')
df1 = df.iloc[:, 1:] # 1열부터 마지막 열까지 선택

df1[df1['temp'].isna()] # 'temp' 온도가 결측치인 행을 선택
df1[df1['mean_wind'].isna()] # 'mean_wind' 강수량이 결측치인 행을 선택
df1[df1['max_wind'].isna()] # 'max_wind' 강수량이 결측치인 행을 선택
```

```python
print(df1.isnull().sum()) # 각 열의 결측치 개수를 출력

# DataFrame.dropna( axis=0, how='any', inplace=False)
# axis=0 : 행을 기준으로 결측치를 제거, axis=1 : 열을 기준으로 결측치를 제거
# how='any' : 하나라도 결측치가 있으면 제거, how='all' : 모두 결측치인 경우에만 제거
# inplace=True : 원본 데이터프레임에 반영, inplace=False : 원본 데이터프레임은 그대로 두고 새로운 데이터프레임을 반환

print(df1.count()) # 결측치가 있는 행의 개수를 출력

df2 = df1
df2.dropna() # 결측치가 있는 행을 제거
print(df2.count()) # 결측치가 제거된 행의 개수를 출력
```

```python
df3 = df1.fillna(df1.mean()) # 결측치를 평균값으로 대체
df3.isnull().sum() # 결측치가 있는 행의 개수를 출력
```

```python
dust = pd.read_excel('data/dust1.xlsx')
dust.head(5)
```

```python
dust1 = dust.drop(['지역', '망', '측정소코드'], axis=1) # '지역', '망', '측정소코드' 열을 제거
dust1.head(5)
```

```python
# 컬럼 이름 변경
dust1.rename(columns={'측정소명' : 'name',
                      '측정일시' : 'date',
                      '주소' : 'addr'}, inplace=True) # 열 이름을 변경
print(dust1.head(5))
# 칼럼 전체 다 변경
dust1.columns = ['name', 'date','SO2', 'CO','O3', 'NO2' ,'pm10', 'pm25', 'addr'] # 열 이름을 변경
dust1.head(5)
```

```python
dust1['date'] = dust1['date'].astype(str) # 'date' 열(object형)을 문자열로 변환
dust1['date'] = dust1['date'].str.slice(0, 8) # 'date' 열의 앞 8자리만 선택
dust1.info()
dust1['date'] = pd.to_datetime(dust1['date']) # 'date' 열을 datetime형으로 변환
dust1
```

```python
dust1.info() # date 의 타입이 datetime64[ns]로 변환된 것을 확인
```

```python
dust1['year'] = dust1['date'].dt.year # 'year' 열을 추가
dust1['month'] = dust1['date'].dt.month # 'month' 열을 추가
dust1['day'] = dust1['date'].dt.day # 'day' 열을 추가
dust1.head(5)
```

```python
#데이터 병합하기
s1 = pd.read_excel('data/nation.xlsx')
s2 = pd.read_excel('data/code.xlsx')
```

```python
merge_df = pd.merge(s1, s2, on='국적코드')
print(merge_df.head(5)) # '국적코드'를 기준으로 병합
merge_df1 = pd.merge(s1, s2, how='left', left_on='국적코드', right_on='국적코드')
print(merge_df1.head(5)) # '국적코드'를 기준으로 왼쪽 외부 조인
merge_df2 = pd.merge(s1, s2, how='right', left_on='국적코드', right_on='국적코드')
print(merge_df2.head(5)) # '국적코드'를 기준으로 오른쪽 외부 조인
```

```python
print(s1.groupby('국적코드').sum()) # '국적코드'를 기준으로 그룹화하여 합계
print(s1.groupby('국적코드')['입국객수'].sum()) # '국적코드'를 기준으로 그룹화하여 '입국객수'의 합계
```

```python
dust1.groupby(['name', 'date'])['SO2'].mean() # 'name'과 'date'를 기준으로 그룹화하여 'SO2'의 평균값
```

```python
dust1.groupby('date')[['SO2', 'CO']].agg(['mean', 'sum']) # 'date'를 기준으로 그룹화하여 'SO2'와 'CO'의 평균값과 합계 
```

```python
dust1.groupby('date').agg({'SO2': ['mean', 'median'], 'CO': ['sum', 'max']}) # 'date'를 기준으로 그룹화하여 'SO2'와 'CO'의 평균값과 합계
```
