```python
import pandas as pd
import numpy as np
```

```python
happy_life = pd.read_excel('data4/삶의만족도.xlsx')
happy_life.head()
```

```python
happy_health=pd.read_excel('data4/건강.xlsx')
happy_health.head()
```

```python
happy_safe=pd.read_excel('data4/안전.xlsx')
happy_safe.head()
```

```python
happy_environ= pd.read_excel('data4/환경.xlsx')
happy_environ.head()
```

```python
happy_econo=pd.read_excel('data4/경제.xlsx')
happy_econo.head()
```

```python
happy_edu=pd.read_excel('data4/교육.xlsx')
happy_edu.head()
```

```python
happy_relation=pd.read_excel('data4/관계 및 사회참여.xlsx')
happy_relation.head()
```

```python
happy_leisure=pd.read_excel('data4/여가.xlsx')
happy_leisure.head()
```

```python
city=list(happy_life['시도'].unique())
happy_merge=pd.DataFrame({'시도':city})
happy_merge
```

```python
life = happy_life['삶의 만족도'].groupby(happy_life['시도']).mean()
life = happy_life.groupby('시도')['삶의 만족도'].mean()
# 둘다 같음
print(life)
```

```python
life = happy_life.groupby('시도')['삶의 만족도'].mean()
happy_merge=pd.merge(happy_merge, life, on='시도')

health = happy_health['평균'].groupby(happy_health['시도']).mean()
happy_merge = pd.merge(happy_merge, health.rename('건강'), on='시도')

safe = happy_safe['평균'].groupby(happy_safe['시도']).mean()
happy_merge = pd.merge(happy_merge, safe.rename('안전'), on='시도')

environ = happy_environ['평균'].groupby(happy_environ['시도']).mean()
happy_merge = pd.merge(happy_merge, environ.rename('환경'), on='시도')

econo = happy_econo['평균'].groupby(happy_econo['시도']).mean()
happy_merge = pd.merge(happy_merge, econo.rename('경제'), on='시도')

edu = happy_edu['평균'].groupby(happy_edu['시도']).mean()
happy_merge = pd.merge(happy_merge, edu.rename('교육'), on='시도')

relation = happy_relation['평균'].groupby(happy_relation['시도']).mean()
happy_merge = pd.merge(happy_merge, relation.rename('관계 및 사회참여'), on='시도')

leisure = happy_leisure['평균'].groupby(happy_leisure['시도']).mean()
happy_merge = pd.merge(happy_merge, leisure.rename('여가'), on='시도')

happy_merge
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
plt.figure(figsize=(15,6))
items = list(happy_merge.columns[1:])
for item in items:
    chartdata = happy_merge[item]
    plt.plot(chartdata, label=item, marker='o', markersize=5)
plt.title('행복지수')
plt.xticks(range(17), list(happy_merge['시도']), rotation=90)
# plt.xlabel('시도')
plt.ylabel('행복지수')
plt.legend()
plt.show()
```

```python
happy_merge.plot(kind='bar', figsize=(15,20), ylabel='시도', xlabel='수치', 
                 grid=True, subplots=True )
plt.suptitle('행복지수')
plt.xticks(range(17), list(happy_merge['시도']), rotation=90)
plt.legend()
plt.tight_layout(pad=2, h_pad=2)
plt.show()
```

```python
plt.Figure(figsize=(15,13))
plt.title("대한민국 행복지수 상관관계", fontsize=16)
up_mask=np.triu(corr)
sns.heatmap(corr, annot=True, mask=up_mask, cmap='RdYlGn')
```
